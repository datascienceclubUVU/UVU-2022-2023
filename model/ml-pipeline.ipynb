```python
import pandas as pd
```

```python
df = pd.read_parquet('processed_features.parquet.gzip')
df
```

```python
df.info()
```

```python
feature_matrix = df.values
```

```python
import sklearn
from sklearn import cluster
```

```python
model = sklearn.cluster.KMeans(n_clusters=10,max_iter=100,verbose=1,random_state=42)
```

```python
history = model.fit(feature_matrix)
```

```python
history.cluster_centers_
```

```python
df2 = pd.DataFrame(history.cluster_centers_)
df2 = df2.rename(columns={0:'release_year',1:'energy',2:'danceability',3:'tempo',4:'valence',5:'acousticness',6:'track_populatity'})
df2
```
