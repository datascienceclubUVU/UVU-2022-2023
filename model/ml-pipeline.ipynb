```python
import pandas as pd
```

```python
%pip install pyarrow

df = pd.read_parquet('processed_features.parquet.gzip',engine='pyarrow')
df.head
```

```python
df.info()
```

```python
feature_matrix = df.values
```

```python
# sklearn imports
import sklearn
from sklearn.cluster import KMeans
```

```python
# Prepare model
model = KMeans(n_clusters=10,max_iter=100,verbose=1,random_state=42)
```

```python
# Fit model
history = model.fit(feature_matrix)
```

```python
history.cluster_centers_
```

```python
# Prepare dataframe
df2 = pd.DataFrame(history.cluster_centers_)
df2 = df2.rename(columns={0:'release_year',1:'energy',2:'danceability',3:'tempo',4:'valence',5:'acousticness',6:'track_populatity'})
df2
```

```python
# Necessary imports

%pip install yellowbrick

from sklearn import datasets
import matplotlib.pyplot as plt
from yellowbrick.cluster import KElbowVisualizer
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
# KElbow Visualizer

km = KMeans(random_state=42)
visualizer = KElbowVisualizer(km,k=(2,10))

visualizer.fit(df2.values)
visualizer.show()
```

```python
# Silhouette Visualizer

fig,ax = plt.subplots(4,2,figsize=(15,8))
for i in [2,3,4,5,6,7,8,9]:
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i,2)

    visualizer = SilhouetteVisualizer(km,colors='yellowbrick',ax=ax[q-1][mod])
    visualizer.fit(df2.values)
```
